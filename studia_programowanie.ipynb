```python
def calculate_product_or_sum(liczba1, liczba2):
    product = liczba1 * liczba2
    if product <= 1000:
        return product
    else:
        return liczba1 + liczba2

# Test Case 1
liczba1_1 = 20
liczba2_1 = 30
result_1 = calculate_product_or_sum(liczba1_1, liczba2_1)
print(f"Biorąc pod uwagę 1:\nliczba1 = {liczba1_1}\nliczba2 = {liczba2_1}\noczekiwany wynik:\nwynik to {result_1}")

print("\n" + "-"*20 + "\n")

# Test Case 2
liczba1_2 = 40
liczba2_2 = 30
result_2 = calculate_product_or_sum(liczba1_2, liczba2_2)
print(f"Biorąc pod uwagę 2:\nliczba1 = {liczba1_2}\nliczba2 = {liczba2_2}\noczekiwany wynik:\nwynik to {result_2}")
```

```python
previous_num = 0

print("Iteracja | Bieżąca Liczba | Poprzednia Liczba | Suma")
print("-------- | -------------- | ---------------- | ----")

for i in range(10): # Iteruje przez liczby od 0 do 9 (pierwsze 10 liczb)
    current_num = i
    current_sum = current_num + previous_num
    print(f"{i+1:<8} | {current_num:<14} | {previous_num:<16} | {current_sum}")
    previous_num = current_num
```

```python
# Zadanie 1: Sprawdź czy pierwsza i ostatnia liczba na liście są takie same.
def check_first_last(num_list):
    if not num_list:
        return False # Lub inna obsługa pustej listy, np. rzucenie wyjątku
    return num_list[0] == num_list[-1]

# Przykłady użycia:
list1 = [10, 20, 30, 10]
print(f"Lista: {list1} -> Pierwsza i ostatnia liczba są takie same: {check_first_last(list1)}")

list2 = [10, 20, 30, 40]
print(f"Lista: {list2} -> Pierwsza i ostatnia liczba są takie same: {check_first_last(list2)}")

list3 = [5]
print(f"Lista: {list3} -> Pierwsza i ostatnia liczba są takie same: {check_first_last(list3)}")

list4 = []
print(f"Lista: {list4} -> Pierwsza i ostatnia liczba są takie same: {check_first_last(list4)}")

print("\n" + "-"*30 + "\n")

# Zadanie 2: Wyświetlanie liczb podzielnych przez 5.
def display_divisible_by_5(num_list):
    print("Liczby podzielne przez 5 z listy:")
    found = False
    for num in num_list:
        if num % 5 == 0:
            print(num)
            found = True
    if not found:
        print("Brak liczb podzielnych przez 5 w tej liście.")

# Przykłady użycia:
list_for_divisible = [10, 23, 35, 42, 50, 7]
display_divisible_by_5(list_for_divisible)

print("\n")

list_no_divisible = [1, 2, 3, 4, 6]
display_divisible_by_5(list_no_divisible)
```

```python
def is_palindrome_number(number):
    # Konwertujemy liczbę na string, aby łatwiej ją odwrócić
    num_str = str(number)
    # Odwracamy string
    reversed_num_str = num_str[::-1]
    # Porównujemy oryginalny string z odwróconym
    return num_str == reversed_num_str

# Przykładowe liczby do zadania:
liczba1 = 90009
liczba2 = 989
liczba3 = 141

print(f"Liczba {liczba1} jest palindromem: {is_palindrome_number(liczba1)}")
print(f"Liczba {liczba2} jest palindromem: {is_palindrome_number(liczba2)}")
print(f"Liczba {liczba3} jest palindromem: {is_palindrome_number(liczba3)}")
```

```python
def reverse_digits_with_spaces(number):
    # Konwertujemy liczbę na string
    num_str = str(number)
    # Odwracamy string
    reversed_str = num_str[::-1]
    # Łączymy znaki spacją
    result = " ".join(reversed_str)
    return result

# Przykład użycia:
liczba_przykladowa = 7536
print(f"Jeśli podana liczba całkowita to {liczba_przykladowa}, wynik będzie wyglądał następująco: \"{reverse_digits_with_spaces(liczba_przykladowa)}\"")
```

```python
list1 = [10, 20, 25, 30, 35]
list2 = [40, 45, 60, 75, 90]

new_list = []

for num in list1:
    if num % 2 != 0:  # Sprawdź, czy liczba jest nieparzysta
        new_list.append(num)

for num in list2:
    if num % 2 == 0:  # Sprawdź, czy liczba jest parzysta
        new_list.append(num)

print(f"Oryginalna lista 1: {list1}")
print(f"Oryginalna lista 2: {list2}")
print(f"Nowa lista zawierająca liczby nieparzyste z list1 i parzyste z list2: {new_list}")
```

```python
# Inicjalizacja pierwszych dwóch wyrazów ciągu Fibonacciego
a, b = 0, 1
count = 0

# Liczba wyrazów do wygenerowania
num_terms = 15

print("Ciąg Fibonacciego:")
while count < num_terms:
    print(a, end=" ")
    nth = a + b
    # Aktualizacja wartości dla kolejnych iteracji
    a = b
    b = nth
    count += 1
print()
```
